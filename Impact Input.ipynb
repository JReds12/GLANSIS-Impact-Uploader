```python
# DO NOT EDIT
import pandas as pd              # Managing dataframes
import os                        # Setting working directores
import time                      # Creates pauses allowing webpage to load

from selenium import webdriver                                     # automate web browser interaction
from selenium.webdriver.common.keys import Keys                    # automate keyboard actions
from selenium.webdriver.common.by import By                        # find elements by html id on webpage
from selenium.webdriver.support.ui import Select                   # automate dropdown selection
from selenium.webdriver.chrome.options import Options              # option to make webdriver not visible

from webdriver_manager.chrome import ChromeDriverManager
```

```python
# Personal NAS username and password
dat = pd.read_excel(r'tech-memo-parasites-update-part2-edit.xlsx', dtype = 'str')
dat['study.location'].fillna('N/A', inplace=True)
dat = dat[48:]
dat
```

```python
# DO NOT EDIT

# Open web page using Selenium webdriver
driver = webdriver.Chrome()
driver.get('https://nas.er.usgs.gov/Dataentry/Impact/ImpactsList.aspx')
```

```python
# new code
for index, row in dat.iterrows():       
    
    # Enter Species ID
    species_id = driver.find_element(By.XPATH, '/html/body/form/div[4]/table[2]/tbody/tr/td[1]/table/tbody/tr[1]/td[2]/input')
    species_id.clear()
    species_id.send_keys(str(row['species.id']))
    
    # Click on 'Add New Impact for Species' Button
    new_impact_button = driver.find_element(By.XPATH, '/html/body/form/div[4]/input[2]')
    new_impact_button.click()
    
    time.sleep(1)
    
    # Input Reference Number
    reference_num = driver.find_element(By.XPATH, '/html/body/form/div[4]/div/div/table[2]/tbody/tr[1]/td[2]/input')
    reference_num.send_keys(str(row['reference']))
    
    time.sleep(1)

    # Find dropdown elements on the web page and select information - Impact Type
    impact_type_dropdown = Select(driver.find_element(By.XPATH, '/html/body/form/div[4]/div/div/table[2]/tbody/tr[2]/td[2]/select'))
    impact_type_dropdown.select_by_visible_text(str(row['impact.type']))
    
    time.sleep(1)
    
    # Find dropdown elements on the web page and select information - Study Type
    study_type_dropdown = Select(driver.find_element(By.XPATH, '/html/body/form/div[4]/div/div/table[2]/tbody/tr[4]/td[2]/select'))
    study_type_dropdown.select_by_visible_text(str(row['study.type']))
    
    time.sleep(1)
    
    # Find dropdown elements on the web page and select information - Study Location
    study_loc_dropdown = Select(driver.find_element(By.XPATH, '/html/body/form/div[4]/div/div/table[2]/tbody/tr[5]/td[2]/select'))
    study_loc_dropdown.select_by_visible_text(str(row['study.location']))
    
    time.sleep(1)

    # Input impact description
    impact_description = driver.find_element(By.XPATH, '/html/body/form/div[4]/div/div/table[2]/tbody/tr[6]/td[2]/textarea')
    impact_description.send_keys(str(row['impact.description']))
    # impact_description.send_keys(str(row['gpt.edit']))
    
    time.sleep(1)
    
    # Input notes
    if pd.notna(row['notes']):
        notes_description = driver.find_element(By.XPATH, '/html/body/form/div[4]/div/div/table[2]/tbody/tr[7]/td[2]/textarea')
        notes_description.send_keys(str(row['notes']))
    
    time.sleep(1)
    
    # Input Great Lakes Region Checkbox
    if row['great.lakes.region'] == "yes":
        gl_region_checkbox = driver.find_element(By.XPATH, '/html/body/form/div[4]/div/div/table[2]/tbody/tr[8]/td[2]/input')
        gl_region_checkbox.click()
    
    time.sleep(1)
    
    # Input Geographic location
    if pd.notna(row['geographic.location']):
        geo_loc = driver.find_element(By.XPATH, '/html/body/form/div[4]/div/div/table[2]/tbody/tr[9]/td[2]/input')
        geo_loc.send_keys(str(row['geographic.location']))
    
    time.sleep(1)

    
    # Assuming you have a list of dataframes named 'dataframes_list'
    tsn_columns_names = [col for col in dat.columns if col.startswith('tsn.')]
    tsn_columns = dat[tsn_columns_names]
    num_columns = tsn_columns.shape[1]
    
    for i in range(1, num_columns):  # Process up to 'tsn.12' if available
        tsn_column = f'tsn.{i}'
        if tsn_column in dat.columns and pd.notna(row[tsn_column]):
            tsn_input = driver.find_element(By.XPATH, '/html/body/form/div[4]/div/div/table[2]/tbody/tr[1]/td[3]/table/tbody/tr[3]/td[2]/input')
            tsn_input.clear()
            tsn_input.send_keys(str(row[tsn_column]))

            add_button = driver.find_element(By.XPATH, '/html/body/form/div[4]/div/div/table[2]/tbody/tr[1]/td[3]/table/tbody/tr[4]/td[2]/input')
            add_button.click()

            time.sleep(1)
    
    # Submit
    submit_button = driver.find_element(By.XPATH, '/html/body/form/div[4]/div/div/div[2]/input')
    submit_button.click()
    
    time.sleep(3)
    
# Close webdriver
#driver.close() 
```
