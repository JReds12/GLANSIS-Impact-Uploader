## upload libraries

```python
import openai
import pandas as pd
import os
import time
```

```python
os.environ["OPENAI_API_KEY"] = ""
```

```python
# import excel sheet

# set working directory
setwd = r''
os.chdir(setwd)

# upload csv with tech memo references
impacts = pd.read_excel("tech-memo-parasites-update2-part1.xlsx")
#impacts = impacts[impacts['gpt.edit'].isna()]
impacts
```

```python
###########################################################################
```

```python
def gpt_response(prompt):
    response = openai.ChatCompletion.create(
      model = "gpt-3.5-turbo",
      temperature = 0,
      request_timeout = 180,
      messages = [
          {"role": "user", "content": prompt},
      ])
    message = response.choices[0]['message']['content']
    return message.strip()
```

```python
impacts["gpt.edit"] = ''

retries = 100 

while retries > 0:
    try:
        for index, row in impacts.iterrows():
            text = "Correct the grammar and syntax of following text and addd <em> HTML tags around any scientific names (binomial nomenclature) in the following text: " + str(row["impact.description"])
            impacts.loc[index, "gpt.edit"] = gpt_response(text)
            print(impacts.loc[index, "gpt.edit"])
            
    except Exception as e:    
        if e: 
            print(e)   
            print('Timeout error, retrying...')    
            retries -= 1    
            time.sleep(5)    
        else:    
             raise e
```
